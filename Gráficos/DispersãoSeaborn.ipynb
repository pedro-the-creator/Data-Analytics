```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('co2.csv')
base.head()
```

```python
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.show()
```

```python
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
plt.figure(figsize=(10, 5))  # Ajuste o tamanho se necessrio
plt.subplot(1, 2, 1)
srn.scatterplot(x=q.conc, y=q.uptake)
plt.title('Quebec')
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.subplot(1, 2, 2)
srn.scatterplot(x=m.conc, y=m.uptake)
plt.title('Mississippi')
plt.tight_layout()
plt.show()
```

```python
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
srn.scatterplot(x=ch.conc, y=ch.uptake)
plt.title('Chilled')
plt.subplot(1, 2, 2)
srn.scatterplot(x=nc.conc, y=nc.uptake)
plt.title('Non chilled')
plt.tight_layout()
plt.show()
```
